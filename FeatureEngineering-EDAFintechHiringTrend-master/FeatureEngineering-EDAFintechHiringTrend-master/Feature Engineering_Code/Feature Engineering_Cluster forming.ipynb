```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
from openpyxl.workbook import Workbook
import re
import csv
import numpy as np
```

```python
joblist_file = pd.read_excel('24-banks-jobs.xlsx')
keyword_file = pd.read_excel('Keyword list(1).xlsx')

payment_cluster_word_list = keyword_file["Payments"].tolist()
wealth_management_cluster_word_list = keyword_file["Wealth Management"].tolist()
blockchain_cluster_word_list = keyword_file["Blockchain"].tolist()
ai_cluster_word_list = keyword_file["Artificial Intelligence"].tolist()
regTech_cluster_word_list = keyword_file["Reg Tech"].tolist()
retail_cluster_word_list = keyword_file["Retail"].tolist()
tech_cluster_word_list = keyword_file["Technology"].tolist()
da_cluster_word_list = keyword_file["Data Analysis"].tolist()
insurance_cluster_word_list = keyword_file["Insurance"].tolist()
```

```python
payment_cluster_word_list = [incom for incom in payment_cluster_word_list if str(incom) != 'nan']
wealth_management_cluster_word_list = [incom for incom in wealth_management_cluster_word_list if str(incom) != 'nan']
blockchain_cluster_word_list = [incom for incom in blockchain_cluster_word_list if str(incom) != 'nan']
ai_cluster_word_list = [incom for incom in ai_cluster_word_list if str(incom) != 'nan']
regTech_cluster_word_list = [incom for incom in regTech_cluster_word_list if str(incom) != 'nan']
retail_cluster_word_list = [incom for incom in retail_cluster_word_list if str(incom) != 'nan']
tech_cluster_word_list = [incom for incom in tech_cluster_word_list if str(incom) != 'nan']
da_cluster_word_list = [incom for incom in da_cluster_word_list if str(incom) != 'nan']
insurance_cluster_word_list = [incom for incom in da_cluster_word_list if str(incom) != 'nan']
```

```python
df2 = pd.DataFrame(columns=['Job ID','Bank Name'])
```

```python
# Payment cluster
# payment_cluster_word_list
list1=[]
list2=[]
list3=[]
list4=[]
list5=[]
list6=[]
n=joblist_file.shape[0]#The value of the last index
#First for loop is for Looping through each url
#Second for loop is for looping through each cluster words
for k in range(0,n):
    for i in payment_cluster_word_list:
        if type(joblist_file['Description'].iloc[k]) == str:
            if i in joblist_file['Description'].iloc[k]:
                desc = joblist_file['Description'].iloc[k]
                c = len(re.findall(i,desc,flags=re.IGNORECASE))
                list3.append(c)
                list1.append(i)
    list2.append(list1[:])
    list4.append(list3[:])
    list5.append(sum(list3[:]))
    list6.append(list5[-1])
    del list1[:]
    del list3[:]
    del list5[:]
    
df2['Job ID'] = joblist_file['Job ID']
df2['Bank Name'] = joblist_file['Bank Name']
df2['cluster']='Payments'
df2['cluster_words']=list2
df2['count'] = list4
df2['weight']=list6
```

```python
# wealth_management_cluster_word_list
# Wealth Management
list1=[]
list2=[]
list3=[]
list4=[]
list5=[]
list6=[]
n=joblist_file.shape[0]#The value of the last index
#First for loop is for Looping through each url
#Second for loop is for looping through each cluster words
for k in range(0,n):
    for i in wealth_management_cluster_word_list:
        if type(joblist_file['Description'].iloc[k]) == str:
            if i in joblist_file['Description'].iloc[k]:
                desc = joblist_file['Description'].iloc[k]
                c = len(re.findall(i,desc,flags=re.IGNORECASE))
                list3.append(c)
                list1.append(i)
    list2.append(list1[:])
    list4.append(list3[:])
    list5.append(sum(list3[:]))
    list6.append(list5[-1])
    del list1[:]
    del list3[:]
    del list5[:]

i=0
for loc in range(0,n+1):
    if i<len(list2):
        df2=df2.append({'cluster_words':list2[i],'cluster':'Wealth Management','Bank Name': joblist_file['Bank Name'].iloc[loc],'Job ID' : joblist_file['Job ID'].iloc[loc], 'count': list4[i],'weight':list6[i]},ignore_index=True)
        i=i+1
```
