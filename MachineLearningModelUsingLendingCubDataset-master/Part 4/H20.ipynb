```python
import h2o
from h2o.estimators.gbm import H2OGradientBoostingEstimator
h2o.init()
```

```python
# Import a sample binary outcome training set into H2O
train = h2o.import_file("featuredEngineering.csv")

# Identify predictors and response
x = train.columns
y = "int_rate"
x.remove(y)
```

```python

train[y] = train[y].asnumeric()

# try using the `keep_cross_validation_predictions` (boolean parameter):
# first initialize your estimator, set nfolds parameter
gbm_model = H2OGradientBoostingEstimator(keep_cross_validation_predictions = True, nfolds = 5, seed = 1)

# then train your model
gbm_model.train(x = x, y = y, training_frame = train)

# print the cross-validation predictions as a list
gbm_model.cross_validation_predictions()

# print the cross-validation predictions as an H2OFrame
gbm_model.cross_validation_holdout_predictions()
```
