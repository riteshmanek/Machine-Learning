```python
# import nlp libraries
import string
import collections
import nltk
from nltk.corpus import stopwords
from nltk.util import ngrams
import nltk, re, string, collections
from nltk.util import ngrams # function for making ngrams

# set nlp variables
english_stops = stopwords.words('english')
lemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
#create functioon to clean text , Lowercases, takes out punct and stopwords and short strings
def cleanning_tokens(words):
    return [words.lower() for words in words if (words not in string.punctuation) and 
                   (words.lower() not in english_stops) and len(words) > 2]

#create function to Lemmatize like remove plurals(play, plays)- it will save- play
def lemmatize(words):
    return [lemmatizer.lemmatize(words) for words in words]
```

```python
#Clean words
clean = cleanning_tokens(words)
#print(clean)
lemmi = lemmatize(clean)
```
