```python
import pandas as pd

df = pd.read_csv('train_lyrics_1000.csv')

df.head()
```

```python
df['genre'].hist() 
df.hist(column="mood",by="genre",bins=30)
```

```python
df['genre'].hist() 
```

```python
import matplotlib.pyplot as plt
import pandas as pd
 
 
pie_chart=df.groupby(df['genre'])['genre'].count()
labels=['Blues','Christian','Country','Dance','Electronic','Hip Hop/Rap', 'Jazz', 'Pop','R&B','Rock','Ska','Soul']
color = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','pink','blue','brown','orange','grey','red','green','purple']
 
plt.pie(pie_chart, labels=labels, colors=color, startangle=90, shadow= False, autopct='%1.1f%%')
 
plt.axis('equal')
plt.show()
```

```python

import matplotlib.pyplot as plt
import pandas as pd

happy=df[df['mood']=='happy']['mood'].count().astype(float)
sad=df[df['mood']=='sad']['mood'].count().astype(float)

x = [2,3]
y = [happy,sad]

bar=plt.bar(x, y, align='center')
bar[0].set_color('b')
bar[1].set_color('g')

plt.title('Mood Distribution')
plt.ylabel('Count')
plt.xlabel('Mood')
plt.xticks(x , ['happy','sad'])

plt.show()

```
